**„ÄêProblem 1] Number of grains of rice on the 100th day**

```python
import numpy as np
import matplotlib.pyplot as plt

list_grains = []
grains = 1
list_grains.append(grains)

for i in range(0,100):
    grains *= 2
    list_grains.append(grains)
    
print("after 100 days grains of rice you receive in total : {}". format(list_grains[len(list_grains)-1]))


plt.title("100 days grains of rice")
plt.xlabel("number of days")
plt.ylabel("number of rice")
plt.xticks(np.arange(1, 100+1,9.0))
plt.tick_params(labelsize=10)
plt.plot(list_grains, 'ro:', linewidth=2, markersize=5) 
plt.show()
```

**„ÄêProblem 2] Number of rice grains outside of the 100th day**

```python
import matplotlib.pyplot as plt

def compute_sorori_shinzaemon(day=100):
    """
    A function that verifies the anecdote of rice by Sorori Shinzaemon. Calculate the number of rice grains per day.
    Parameteres
    --------------
    day: int
        Number of days to get rice (default : 100)
    Returns
    --------------
    list_n_grains : list
        List of the number of rice you get one day (the length of the list is equal to day)
    list_total_grains : list
        List of the total number of rice you will receive by a certain day (the length of the list will be equal to day)
    """
    list_n_grains = []
    list_total_grains = []
    grains = 1
    total = grains
    list_n_grains.append(grains)
    list_total_grains.append(grains)
    for i in range(0,day):
        grains *= 2
        list_n_grains.append(grains)
        total += grains
        list_total_grains.append(total)
    pass
    return list_n_grains, list_total_grains

def show(lists, color='r'):
    plt.title("100 days grains of rice")
    plt.xlabel("number of days")
    plt.ylabel("number of rice")
    plt.xticks(np.arange(0, len(lists),1.0))
    plt.tick_params(labelsize=10)
    plt.plot(lists, color+'o:', linewidth=2, markersize=5) 
    plt.show()
    

list_n_grains, list_total_grains = compute_sorori_shinzaemon(day=10)


print("length n : {} and n : \n".format(len(list_n_grains)))
print(list_n_grains)
show(list_n_grains,'r')

print("\nlenght total : {} and total : \n".format(len(list_total_grains)))
print(list_total_grains)
show(list_total_grains,'b')
```

**„ÄêProblem 3] How many people can live for how many days with the rice we get?**

From my research in website https://www.msn.com/en-in/health/nutrition/rice-and-easy-people-should-consume-an-extra-50g-of-rice-every-day-to-beat-obesity-so-how-much-are-people-eating-in-your-country/ar-AAALTAf
\
I using data number of rice each day in (VietNam, Japan and China)
\
fomular i using is total rice_grains/(pepole* rice_grains_needed_day)

```python
import matplotlib.pyplot as plt
import numpy as np

def how_many_days(rice_grains, pepole, rice_grains_needed_day):
    days_live = rice_grains/(pepole* rice_grains_needed_day)
    pass
    return days_live

def compute_sorori_shinzaemon(day):
    list_total_grains = []
    grains = 1
    total = grains
    list_total_grains.append(grains)
    for i in range(0,day):
        grains *= 2
        total += grains
        list_total_grains.append(total)
    pass
    return list_total_grains

list_total_grains = compute_sorori_shinzaemon(15)

lists = [['Vietnam','China','Japan','Indonesia','Lao'],[398,212.46, 149.02,361.04,443],[]]

for i in range(0, len(lists[0])):
    lists[0][i] += '\n'+ str(lists[1][i]) + 'g'
    pass
    
for i in lists[1]:
    days_live = how_many_days(list_total_grains[len(list_total_grains)-1], 40, i)
    lists[2].append(days_live)
    pass

plt.title("How many days live in seperate Country")
plt.xlabel("Country")
plt.ylabel("Days live")
plt.bar(lists[0],lists[2])
plt.yticks(np.arange(0, max(lists[2]) + 3, 1.0))
for i in range(len(lists[1])):
    plt.text(i, lists[2][i] + 0.5, "{:.2f}".format(lists[2][i]), horizontalalignment='center', verticalalignment='center', size = 10)
plt.show()
```
