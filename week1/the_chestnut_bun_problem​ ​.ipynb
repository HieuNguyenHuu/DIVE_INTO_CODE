**[Problem] The day when chestnut buns cover the solar system**

```python
def show(lists,listsm, color='r', name_object = 'Universe'):
    plt.title("Chestnut buns cover " + name_object)
    plt.xlabel("Number of hours")
    plt.ylabel("Volume of Chestnut buns")
    plt.tick_params(labelsize=10)
    plt.xticks( np.arange(0, listsm[len(listsm) -1 ],12),np.arange(0, len(listsm)) )
    plt.plot(lists, color+ 'o:', linewidth=2, markersize=5) 
    plt.show()

def chestnut_buns(chestnut_volume, object_volume):
    lists_chestnut = [[],[]]
    t0 = chestnut_volume
    minute = 0
    lists_chestnut[0].append(t0)
    lists_chestnut[1].append(minute)
    while t0 <= object_volume:
        t0 *= 2
        minute += 5
        lists_chestnut[0].append(t0)
        lists_chestnut[1].append(minute)
    return lists_chestnut


lists_chestnut_universe = chestnut_buns(0.0138, 3.58 * (10**80))


show(lists_chestnut_universe[0],lists_chestnut_universe[1])

print("After {:.3f} hours the Chestnut will cover over the Universe\n".format(lists_chestnut_universe[1][len(lists_chestnut_universe[1])-1]/60))

print("chestnut buns double in {} times".format(len(lists_chestnut_universe[0])))
print("chestnut buns : {:.3f} in 10^80 m^3 ".format(lists_chestnut_universe[0][len(lists_chestnut_universe[0])-1] / (10**80)))
print("Universe : {:.3f} in 10^80 m^3".format(3.58 ))
```

```python

def show(lists,listsm, color='r', name_object = 'Universe'):
    plt.title("Chestnut buns cover " + name_object)
    plt.xlabel("Number of hours")
    plt.ylabel("Volume of Chestnut buns")
    plt.tick_params(labelsize=10)
    plt.xticks( np.arange(0, listsm[len(listsm) -1 ],12),np.arange(0, len(listsm)) )
    plt.plot(lists, color+ '>:', linewidth=2, markersize=5) 
    plt.show()
    print("After {:.3f} hours the Chestnut will cover over the {}".format(listsm[len(listsm)-1]/60, name_object))
    print("chestnut buns double in {} times".format(len(listsm)))
    
def chestnut_buns(chestnut_volume, object_volume):
    lists_chestnut = [[],[]]
    t0 = chestnut_volume
    minute = 0
    lists_chestnut[0].append(t0)
    lists_chestnut[1].append(minute)
    while t0 <= object_volume:
        t0 *= 2
        minute += 5
        lists_chestnut[0].append(t0)
        lists_chestnut[1].append(minute)
    return lists_chestnut

lists_object = [['Tokyo Dome', 'Mt Fuju', 'Moon' , 'Mars'],[1.24*(10**6), 400*(10**9),2.1968*(10**10), 16.318*(10**10)]]

lists_color = ['r','b','c','m']

lists = [[],[],[],[]]

listsminutes = []

chestnut_volume = 0.0138

j=0
for i in lists_object[1]:
    liststemp = chestnut_buns(chestnut_volume, i)
    lists[j].append(liststemp)
    j+=1
    pass

j=0
for i in lists:
    show(i[0][0], i[0][1],lists_color[j], lists_object[0][j])
    listsminutes.append(i[0][1][len(i[0][1]) - 1])
    j+=1

plt.title("Minutes seperate Object")
plt.xlabel("Object")
plt.ylabel("Minutes")
plt.bar(lists_object[0],listsminutes)
plt.yticks(np.arange(0, max(listsminutes)+50,50))
for i in range(len(listsminutes)):
    plt.text(i, listsminutes[i] + 8, "{}m".format(listsminutes[i]), horizontalalignment='center', verticalalignment='center', size = 10)
plt.show()
```
