**[Problem 1] Create using the exponentiation arithmetic operator**

```python
THICKNESS = 0.00008
folded_thickness = THICKNESS*(2**43)
print("Thickness: {} meters".format(folded_thickness))
```

**[Problem 2] Unit Conversion**

```python
print("Thickness: {: .2f} in 10,000 kilometers".format(10000*folded_thickness/1000))
```

**[Problem 3] Create using a for statement**

```python
folded_thickness = THICKNESS
for i in range(0, 43):
    folded_thickness *= 2
print("Thickness: {: .2f} in 10,000 kilometers".format(10000*folded_thickness/1000))
```

**[Problem 4] Comparison of calculation time**

```python
import time
start = time.time()

THICKNESS = 0.00008
folded_thickness = THICKNESS
for i in range(0, 43):
    folded_thickness *= 2
print("Thickness in For Method: {: .2f} in 10,000 kilometers".format(10000*folded_thickness/1000))

elapsed_time = time.time() - start
print("\ntime for For Method : {}[s]\n".format(elapsed_time))

import time
start = time.time()

THICKNESS = 0.00008
folded_thickness = THICKNESS*(2**43)
print("Thickness in Exponentiation Method: {: .2f} in 10,000 kilometers".format(10000*folded_thickness/1000))

elapsed_time = time.time() - start
print("\ntime for Exponentiation Method : {}[s]".format(elapsed_time))
```

The better is Exponentiation Method with time is shorter than For Method

**[Problem 5] Saving to a list**

```python
List_value_folds = []
folded_thickness = THICKNESS
List_value_folds.append(folded_thickness)

for i in range(0, 43):
    folded_thickness *= 2
    List_value_folds.append(folded_thickness)
    
print("length of List contain value of folds : {}".format(len(List_value_folds)))
```

**[Problem 6] Displaying a line graph**

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(List_value_folds) 
plt.show()
```

The thickness of folded paper have proliferated in folds at 35 times, while the thickness is just change smaller in the number of folds befor 30 times

**[Problem 7] Customizing graphs**

```python
plt.title("thickness of folded paper first customized")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.tick_params(labelsize=10)
plt.plot(List_value_folds,'--', color='yellow', linewidth=2, markersize=12)
plt.show()
```

```python
plt.title("thickness of folded paper second customized")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.tick_params(labelsize=10)
plt.plot(List_value_folds,'o--', color='#ffb3e6', linewidth=3, markersize=12)
plt.show()
```

```python
plt.title("thickness of folded paper thirds customized")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.tick_params(labelsize=10)
plt.plot(List_value_folds,'D:', color='#de6c60', linewidth=3, markersize=10)
plt.show()
```
