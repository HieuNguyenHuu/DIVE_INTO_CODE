**[Problem 6] Creating a diagram**

<hr>

**A pie chart of the number of samples per label (while showing percentages)**

```python
df2 = pd.DataFrame({'Species' : ['setosa', 'versicolor', 'virginica'],
                    'Pecentage': (df['Species'].value_counts()/df['Species'].sum()) *100})

df2.set_index('Species')

plot = df2.plot.pie(title= 'the number of samples per label', y='Pecentage',labels = df2['Species'], figsize=(5, 5), autopct='%1.1f%%', startangle=90, labeldistance=0.5,pctdistance=0.7, wedgeprops={'linewidth': 2, 'edgecolor':"white"}, counterclock=False)
```

**Select features one by one and visualize the distribution of data for each label using a box plot**

```python
df.loc[df['Species'] == 0].boxplot(column = list(df.columns.values[:-1]))
plt.title('Box plot for Species \"Setosa\"')
plt.show()

df.loc[df['Species'] == 1].boxplot(column = list(df.columns.values[:-1]))
plt.title('Box plot for Species \"Versicolor\"')
plt.show()

df.loc[df['Species'] == 2].boxplot(column = list(df.columns.values[:-1]))
plt.title('Box plot for Species \"Virginica\"')
plt.show()
```

**Select features one by one and visualize the distribution of data for each label using a violin plot**

```python
lists = list(df.columns.values[:-1])

for k in lists:
    plt.violinplot([df.loc[df["Species"]==0][k],df.loc[df["Species"]==1][k],df.loc[df["Species"]==2][k]])
    plt.xticks([1,2,3] , ['setosa', 'versicolor', 'virginica'])
    plt.xlabel("Species")
    plt.ylabel("Number")
    plt.title('Violin plot for {}'.format(k))
    plt.show()
```

**different between violin plot and box plot**

* Violin plots contain more information than box plots. The boxplot contains only the mean / median and interquartile range summary statistics, while the violin plot contains all distribution data

* The violin plot is less important than the boxplot, despite the fact that it contains more detail. It is difficult for people who are unfamiliar with the violin plot to grasp the context due to its low name. It is possible to draw a plurality of histograms or the kernel density as an alternative diagram that is easier to understand in this situation. The quartile is easily visible in the box map. The violin plot depicts the distribution and is simple to understand.

**Take one feature for each of the vertical and horizontal axes and create a color-coded scatter plot for each type (6).***

```python
sepal = ['sepal_length','sepal_width']

petal = ['petal_length','petal_width']

spec = d.target_names

for i in range(len(spec)):  
    plt.scatter(df.loc[df["Species"]==i][sepal[0]],df.loc[df["Species"]==i][sepal[1]],label=spec[i])

plt.title("Sepal scatter plot")
plt.xlabel(sepal[0])
plt.ylabel(sepal[1])
plt.legend()
plt.show()

for i in range(len(spec)):  
    plt.scatter(df.loc[df["Species"]==i][petal[0]],df.loc[df["Species"]==i][petal[1]],label=spec[i])
    
plt.title("Petal scatter plot")
plt.xlabel(petal[0])
plt.ylabel(petal[1])
plt.legend()
plt.show()

```
