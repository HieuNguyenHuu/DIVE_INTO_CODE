**[Problem 1] Number of wheat on a 2 x 2 square chess board**

<hr>

```python
n_squares = 4
small_board_list = [1]
for i in range(n_squares - 1):
    small_board_list.append(2*small_board_list[-1])

small_board_ndarray = np.array(small_board_list)

square_array = small_board_ndarray.reshape(2,2)

for i,value in enumerate(square_array):
    for j, value in enumerate(value):
        print("number of wheat in square [{},{}] : {}".format(j+1,i+1,value))


```

**[Problem 2] Expansion to n Ã— m mass**

<hr>

```python
def mxn_mass(m, n):
    t0 = 1 
    lists = [t0]
    for i in range(m*n - 1):
        lists.append(lists[-1]*2) 
    arr = np.array(lists)
    arr = arr.reshape(m,n)
    
    return arr

arr = mxn_mass(2, 3)

print("run function in simple example 2x3 for checking : ")
print(arr.sum())
print("\n")


arr = mxn_mass(8, 8)

for i,value in enumerate(arr):
    for j, value in enumerate(value):
        print("number of wheat in square [{},{}] : {}".format(j,i,int(value)))
```

**[Problem 6] Expansion to n Ã— m mass by another calculation method**

<hr>

```python
def mxn_append_method(m, n):
    arr = np.array([1]).astype(np.uint64)
    for i in range(m*n - 1):
        arr = np.append(arr, 2*arr[-1]) 
    arr = arr.reshape(m,n)
    return arr


def mxn_broadcast_method(m, n):
    lists = np.arange(m*n).astype(np.uint64)
    arr = 2**lists
    arr = arr.reshape(m,n)
    return arr

arr_append = mxn_append_method(8,8)

print("in Append method 8x8 square is \n{}\n".format(arr_append))

arr_broadcast = mxn_broadcast_method(8,8)

print("in Broadcase method 8x8 square is \n{}\n".format(arr_append))
    
```
