# Mathematics for Machine Learning Pre-Class assignment 2 Creating Dummy Data

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
np.random.seed(0)
```

<hr>

**[Problem 1] Creating random numbers** 

<hr>


```python
mean = (-3, 0)
cov = [[1.0, 0.8], [0.8, 1.0]]
x = np.random.multivariate_normal(mean, cov, 500)
print("The shape of x is : {}".format(x.shape))

```

<hr>

**[Problem 2] Visualization by scatter plot**

<hr>

```python
x1,x2 = x.T
plt.title("Scatter of x")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(x1 ,x2)
plt.show()
```

<hr>

**[Problem 3] Visualization by histogram**

<hr>

```python
plt.title("Histogram of x")
plt.xlabel("x1")
plt.ylabel("x2")
plt.hist(x)
plt.show()


plt.title("Histogram of x1")
plt.xlabel("x1")
plt.ylabel("frequency")
plt.hist(x1)
plt.xlim(-5,2)
plt.show()


plt.title("Histogram of x2")
plt.xlabel("x2")
plt.ylabel("frequency")
plt.hist(x2)
plt.xlim(-5,2)
plt.show()
```

<hr>

**[Problem 4] Addition of data**

<hr>

```python
mean = (0, -3)
cov = [[1.0, 0.8], [0.8, 1.0]]
xx = np.random.multivariate_normal(mean, cov, 500)
print("The shape of new xx is : {}\n".format(x.shape))

x1,x2 = x.T
xx1, xx2 = xx.T

plt.title("Scatter of x (0) and xx (1)")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(x1 ,x2)
plt.scatter(xx1 ,xx2)
plt.legend([0,1])
plt.show()
```

<hr>

**[Problem 5] Data combination**

<hr>

```python
x_combination = np.concatenate((x, xx), axis = 0)

print("the shape of combination of x and xx is : {}".format(x_combination.shape))
```

<hr>

**[Problem 6] Labeling**

<hr>


```python
x = np.concatenate((x, np.array([[0] * 500]).T), axis=1)
print("new x with lable 0 shape {} \n{}".format(x.shape,x)) 
print("\n")
xx = np.concatenate((xx, np.array([[1] * 500]).T), axis=1)
print("new xx with lable 1 shape {} \n{}".format(xx.shape,xx)) 
print("\n")

x_combination = np.concatenate((x, xx), axis = 0)

print("the shape of combination of x and xx is : {}".format(x_combination.shape))

print(x_combination)
```
