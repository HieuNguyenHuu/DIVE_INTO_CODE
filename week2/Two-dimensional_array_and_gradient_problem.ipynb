# Mathematics lesson for machine learning assignment 1 Two-dimensional array and gradient problem

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import math  
```

<hr>

**[Problem 1] Linear function**

<hr>

```python
x_adarry =  np.arange(-50, 50.1, 0.1)

y_adarry =  1 / 2 * x_adarry  + 1

print("x array arange (-50, 50) {} \n\n{}\n".format(x_adarry.shape,x_adarry))

print("y array with formula (y = 1/2x + 1) \n\n{}".format(y_adarry))
```

<hr> 

**[Problem 2] Array combination**

<hr>

```python
n = np.concatenate((x_adarry.reshape(len(x_adarry), 1), y_adarry.reshape(len(y_adarry), 1)), axis = 1)

print("Array combination \n\n{}\n".format(n))
```

<hr>

**[Problem 3] Find the gradient**

<hr>

```python
dx = n[:,0][1:] - n[:,0][:-1]

dy = n[:,1][1:] - n[:,1][:-1]

gradient = dy / dx

print("gradient shape {} and : \n{}".format(gradient.shape, gradient))
```

<hr>

**[Problem 4] Draw a graph**

<hr>

```python

plt.title("Two-dimensional array and gradient problem y=1/2*x+1")
plt.xlabel("X") 
plt.ylabel("Y") 
plt.plot(x_adarry, y_adarry)
plt.plot(x_adarry[1:], gradient)
plt.scatter(x_adarry[np.argmin(y_adarry)],np.min(y_adarry))
plt.legend(["y=1/2*x+1", "Gradient", "min value"])
plt.show()
```

**[Problem 5] Python functionalization**

<hr>

```python
def compute_gradient(function, x_range = (-50, 50.1, 0.1)):
    xtemp = np.arange(*x_range)
    ytemp = function(xtemp)
    array_xy = np.concatenate((xtemp.reshape(xtemp.shape[0],1), ytemp.reshape(ytemp.shape[0],1)), axis = 1)
    dx = array_xy[:,0][1:] - array_xy[:,0][:-1]
    dy = array_xy[:,1][1:] - array_xy[:,1][:-1]
    gradient =  dy/dx
    pass
    return array_xy, gradient

def function1(array_x):
    array_y = array_x**2
    pass
    return array_y

def function2(array_x):
    array_y = 2 * (array_x**2) + (2**array_x)
    pass
    return array_y

def function3(array_x):
    array_y = np.sin((array_x** (1/2)))
    pass
    return array_y

def show(array_xy, gradient, name_function):
    plt.title("Two-dimensional array and gradient problem \n" +  name_function)
    plt.xlabel("X") 
    plt.ylabel("Y") 
    plt.plot(array_xy[:,0], array_xy[:,1])
    plt.plot(array_xy[:,0][1:], gradient)
    plt.legend([ name_function, "Gradient"])
    plt.show()

array_xy, gradient = compute_gradient(function1)

show(array_xy, gradient, "y=x^2")

array_xy, gradient = compute_gradient(function2)

show(array_xy, gradient, "y=2x^2+2^x")

array_xy, gradient = compute_gradient(function3, (0, 50.1, 0.1))

show(array_xy, gradient, "y=sin(x^(1/2))")
```
